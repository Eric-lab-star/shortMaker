# Models 
LLM은 Large Language Model의 약자로서 인공지능의 두뇌에 해당하는 강력한 도구이다. LLM의 구체적인 기능으로는 외부 데이터 API를 호출하는 등의 도구를 상용하고 사용자가 지정한 형식으로 답변하는 것이 가능하며 문장을 생성하는 것을 넘어 이미지, 동영상, 오디오 생성 또한 가능하며 결론에 도달하기까지 여러 단계를 거처 추론하는 것이 가능하다. 

LLM의 성능은 에이전트의 수행능력에 큰 영향을 준다. 어떤 모델은 복잡한 지침을 잘 수행하며, 어떤 모델은 간단 지침을 빠르게 수행한다. Langchain은 다양한 모델을 쉽게 바꾸어 가며 사용할 수 있도록 설계되어있다. 


## Model 생성방법

```python
import dotenv
from langchain.chat_models import init_chat_model

dotenv.load_dotenv()

model = init_chat_model("claude-sonnet-4-6")
```

```python
response = model.invoke("Why do parrots talk?")
```

```python
response.content
```

### Model 생성 인수

*model*

- 사용하고 싶은 모델의 이름. 모델의 이름은 "[공급자]:[이름]" 형식으로 작성한다.

*api_key*

- 모델공급자의 모델을 사용하기 위한 api키. 일반적으로 환경변수에 저장해서 사용한다. 

*temperature*

- 숫자가 높으면 답변이 더 창의적이로 답변하고 숫자가 낮으면 더 결정론적으로 답을 한다.

*max_tokens*

- 답변을 할 때 사용되는 사용되는 최대 토큰수. 숫자가 높으면 답변하는 시간이 오래걸린다.

*timeout*

- 설정한 시간보다 답변 시간이 오래걸리면 요청을 취소한다. 


```python
model_sonnet = init_chat_model(
    "claude-sonnet-4-6",
    temperature=0.7,
    timeout=30,
    max_tokens=1000,
)
```

### Invoke

#### 1. 단일 메시지

```python
from rich.pretty import pprint
response = model_sonnet.invoke("김경섭이 어떤 사람인지 알려줘")
pprint(response)
```

#### 2. 메시지 리스트

```python
conversation = [
    {
        "role": "system",
        "content": "You are a helpful assistant that translates English to Korean",
    },
    {
        "role": "user",
        "content": "Translate: I love programming",
    },
    {
        "role": "assistant",
        "content": "나는 프로그래밍을 사랑한다.",
    },
    {
        "role": "user",
        "content": "Translate: I love building applications" 
    },
    
]

response = model_sonnet.invoke(conversation)
pprint(response.content)
```

```python
from langchain.messages import HumanMessage, AIMessage, SystemMessage

conversation = [
    SystemMessage("너는 끝말잇기 최강자이다."),
    HumanMessage("사다리"),
    AIMessage("리어카"),
    HumanMessage("카센터"),
]
response = model_sonnet.invoke(conversation)
pprint(response.content)
```

#### stream

모델의 답변시간이 길어지는 경우, 완성된 글자들부터 출력하는 방법으로 사용자 경험을 향상시킨다. 

```python
response = model_sonnet.stream("인공지능의 신경망과 학습과정을 500자 이내로 설명해줘")
for chunk in response:
    print(chunk.text, end="", flush=True)
```

AIMessage를 전달하는 Invoke메소드와 달리, stream 방식은 AIMessageChunk를 반환한다. 

#### Batch 

모델은 요청을 병렬로 처리하기 때문에 여러개의 질문을 한 묶음으로 요청하면 성능향상과, 비용 절감이 가능하다. 

```python
responses = model.batch([
    "Why sky is blue",
    "What is origin of digital sensor?",
    "Who is the richest man in the world?"
])
for response in responses:
    print(response.content)
```

#### Tool Calling 

모델은 도구를 사용해서 외부 데이터 가져오기 인터넷 검색하기 코드 실행하기와 같은 작업을 수행할 수 있다.

### 1. 도구 정의하기

```python
from langchain.tools import tool
from langchain.messages import AIMessage, HumanMessage, SystemMessage

@tool
def get_weather(location: str) -> str:
    """Get the weather at a location."""
    return f"It's sunny in {location}"

@tool
def calculate(expression: str) -> str:
    """수식을 계산합니다. 예 '2+3 = 5' """
    return str(eval(expression))
```

### 2 LLM에 Tool 연결하기

```python

model_with_tools = model.bind_tools([get_weather, calculate])
```

### 3. LLM에게 질문하기

```python
messages: list[HumanMessage|AIMessage| SystemMessage] = [
    HumanMessage("What's the weather like in Boston"),
]

ai_response = model_with_tools.invoke(messages)
```

### 4. tool 직접 실행하기

```python

tool_map = {
    "get_weather": get_weather,
    "calculate": calculate,
}

messages.append(ai_response) # LLM 응답을 대화기록에 추가하기
```

### 5. Tool 직접 실행하기

```python
for tool_call in ai_response.tool_calls:
    tool_name = tool_call["name"]
    tool_args = tool_call["args"]

    tool_result = tool_map[tool_name].invoke(tool_call)
    messages.append(tool_result)

final_response = model_with_tools.invoke(messages)
print(final_response )
```

### Structured output

결과를 구조화 시키면 다음 처리 과정에서 데이터를 다루기 쉬워지는 효과가 있다. 랭체인은 구조화를 시키는 다양한 방법을 제시한다.
